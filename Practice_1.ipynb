```python
# Shuffling the order of a tensor


not_shuffled = tf.constant([[12,3], [30,4], [5, 6]])

# Shuffle our non-shuffled tensor
shuffled = tf.random.shuffle(not_shuffled, seed=42)

shuffled
```

```python
import tensorflow as tf

tensor_1 = tf.constant([
    [3, 4, 5], 
    [6, 6, 8], 
    [3, 3, 3]
], dtype=tf.float64)  # Ensure float64 type if needed

tensor_2 = tf.constant([
    [1, 2], 
    [3, 4], 
    [5, 6]
], dtype=tf.float64)

# Check shapes
print("Shape of tensor_1:", tensor_1.shape)  # (3, 3)
print("Shape of tensor_2:", tensor_2.shape)  # (3, 2)

# Adjust axes for correct dot product
result = tf.tensordot(tensor_1, tensor_2, axes=1)

print("Result shape:", result.shape)  # Expected: (3, 2)
print(result)
```
